```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

print("--- Step 1 & 2: Loading California Housing Dataset ---")
# Dataset load karna
data = fetch_california_housing(as_frame=True)
df = pd.concat([data.data, data.target.rename("HousePrice")], axis=1)
print(df.head())

print("\n--- Step 3: Splitting Features and Target ---")
# Features (X) aur Target (y) alag karna
X = df.drop("HousePrice", axis=1)
y = df["HousePrice"]

print("\n--- Step 4: Applying Feature Scaling ---")
# Data ko standard scale par lana
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Feature scaling successfully completed.")

print("\n--- Step 5: Train-Test Split (80% Train, 20% Test) ---")
# 80/20 train-test split karna
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")

print("\n--- Step 6 & 7: Training Multiple Models & Comparing Results ---")
# Teeno models ko define karna
models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Decision Tree": DecisionTreeRegressor(max_depth=5, random_state=42)
}

# Training aur performance save karne ke liye dictionary
results = {}

for name, model in models.items():
    # Model ko train karna
    model.fit(X_train, y_train)
    # Predictions nikalna
    predictions = model.predict(X_test)

    # Error fixed: squared=False ki jagah np.sqrt(mean_squared_error(...)) use kiya hai
    mse = mean_squared_error(y_test, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, predictions)

    results[name] = {
        "RMSE": round(rmse, 4),
        "R2 Score": round(r2, 4)
    }

# Results ko table form mein convert karna
results_df = pd.DataFrame(results).T
print("\n===== FINAL MODEL PERFORMANCE COMPARISON =====")
print(results_df)
print("==============================================")

print("\n--- Step 8: Visualizing Best Model ---")
# Linear Regression ko validation ke liye plot karna
best_model = LinearRegression()
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)

plt.figure(figsize=(7, 6))
plt.scatter(y_test, y_pred, alpha=0.3, color='teal')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color="red", linestyle="--", linewidth=2)
plt.xlabel("Actual House Prices")
plt.ylabel("Predicted House Prices")
plt.title("Actual vs Predicted House Prices (Task 2 Baseline)")
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()
```
